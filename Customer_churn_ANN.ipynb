```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges,errors='coerce')
```

```python
df.drop('customerID',axis='columns',inplace=True)
```

```python
df
```

```python
# Checking the Non-numerical/ categorical values
for column in df:
    if(df[column].dtypes=='object'):
        print(f'{column}: {df[column].unique()}') 
```

```python
# MultipleLines, OnlineSecurity, OnlineBackup and many columns has no internet services whihc means that customers are not able to opearte that facility, so we can consider those cases unser the same 'No' label
df.replace('No internet service','No',inplace=True)
df.replace('No phone service','No',inplace=True)
```

```python
for column in df:
    if(df[column].dtypes=='object'):
        print(f'{column}: {df[column].unique()}') 
```

```python
# now we can encode the yes:1, no:0 very easily and train the ANN
```

```python
label_encode_columns = []
for column in df:
    if(df[column].dtypes=='object'):
        label_encode_columns.append(column)
        #print(f'{column}: {df[column].unique()}') 
```

```python
label_encode_columns
```

```python
# We won't apply yes no encoding on these
# gender
# InternetService: ['DSL' 'Fiber optic' 'No']
# Contract: ['Month-to-month' 'One year' 'Two year']
# PaymentMethod: ['Electronic check' 'Mailed check' 'Bank transfer (automatic)' 'Credit card (automatic)']
```

```python
label_encode_columns.remove('gender')
label_encode_columns.remove('InternetService')
label_encode_columns.remove('Contract')
label_encode_columns.remove('PaymentMethod')
```

```python
label_encode_columns
```

```python
for i in label_encode_columns:
    df[i].replace({'Yes':1,'No':0},inplace=True)
```

```python
df['gender'].replace({'Female':0,'Male':1},inplace=True)
```

```python
for column in df:
    print(f'{column}: {df[column].unique()}')
```

```python
# InternetService, Contract, PaymentMethod are the columns to which we will apply one hot encoding
```

```python
df = pd.get_dummies(data=df,columns=['InternetService', 'Contract', 'PaymentMethod'])
```

```python
df.columns
```

```python
# removed 3 and added 10 columns so total 20+7 columns
```

```python
df.dtypes
```

```python
df.tenure,df.MonthlyCharges, df.TotalCharges
```

```python
# Scaling tenure, monthlycharges, totalcharges between -1 to 1
from sklearn.preprocessing import MinMaxScaler
scl = MinMaxScaler()
df[['tenure','MonthlyCharges','TotalCharges']] = scl.fit_transform(df[['tenure','MonthlyCharges','TotalCharges']])
```

```python
for column in df:
    print(f'{column}: {df[column].unique()}')
```

```python
df.to_csv('Customer-Churn-Encoded.csv')
```

```python
x = df.drop('Churn',axis=True)
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape, y_train.shape
```

```python
# ANN
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
```

```python
model = keras.Sequential([
    keras.layers.Dense(units = 20, input_shape=(26,), activation='relu'),
    keras.layers.Dense(units = 15, activation='relu'),
    keras.layers.Dense(units = 1, activation='sigmoid')
])
```

```python
model.compile(optimizer = 'adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=50)
```

```python
df.dropna(how='any',inplace=True)
```

```python
df.isnull().sum()
```

```python
# now the null values from the df is removed, 11 rows will be deleted
```

```python
print(df.shape)
x = df.drop('Churn',axis=True)
y = df['Churn']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5)
x_train.shape, x_test.shape
```

```python
# ANN
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
model = keras.Sequential([
    keras.layers.Dense(units = 20, input_shape=(26,), activation='relu'),
    keras.layers.Dense(units = 15, activation='relu'),
    keras.layers.Dense(units = 10, activation='relu'),
    keras.layers.Dense(units = 1, activation='sigmoid')
])
model.compile(optimizer = 'adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=50,validation_split=0.3)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
# the error is occuring because we have got the y_pred as float and 2D value, we need to convert in=t into binary classification and 1D
print(y_pred)
print("\n")
print(y_test)
```

```python
y_pred = y_pred.ravel() # convert ND to 1D
```

```python
for i in range(len(y_pred)):
    if(y_pred[i]>0.5):
        y_pred[i]=1
    else:
        y_pred[i]=0
```

```python
y_pred = y_pred.astype(int)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("accuracy: ",accuracy_score(y_pred,y_test),"\n"), print("classification_report: ",classification_report(y_pred,y_test))
print("confusion_matrix: \n",confusion_matrix(y_pred,y_test))
```
